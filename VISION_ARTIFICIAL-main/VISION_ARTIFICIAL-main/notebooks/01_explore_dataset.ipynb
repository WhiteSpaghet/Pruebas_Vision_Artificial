# VISION_ARTIFICIAL - Notebook 01: Exploración del dataset CIFAR-10

# 🔹 Importar librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import cifar10
```

🔹 Configuración de estilo

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12,8)
```

🔹 Cargar CIFAR-10

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

🔹 Información básica del dataset

```python
print("Tamaño de entrenamiento:", x_train.shape, y_train.shape)
print("Tamaño de test:", x_test.shape, y_test.shape)
```

🔹 Etiquetas de las clases

```python
class_names = ['Avión', 'Coche', 'Pájaro', 'Gato', 'Ciervo',
               'Perro', 'Rana', 'Caballo', 'Barco', 'Camión']
```

🔹 Función para mostrar imágenes aleatorias de cada clase

```python
def mostrar_ejemplos_por_clase(x, y, class_names, ejemplos=5):
    plt.figure(figsize=(20,10))
    for idx, class_name in enumerate(class_names):
        indices = np.where(y.flatten() == idx)[0]
        selected = np.random.choice(indices, ejemplos, replace=False)
        for i, sel in enumerate(selected):
            plt.subplot(len(class_names), ejemplos, idx*ejemplos + i + 1)
            plt.imshow(x[sel])
            plt.axis('off')
            if i == 2:
                plt.title(class_name, fontsize=12)
    plt.tight_layout()
    plt.show()
```

🔹 Mostrar ejemplos

```python
mostrar_ejemplos_por_clase(x_train, y_train, class_names)
```

🔹 Distribución de clases

```python
sns.countplot(y_train.flatten())
plt.xticks(ticks=np.arange(10), labels=class_names, rotation=45)
plt.title("Distribución de clases en el set de entrenamiento")
plt.show()
```

🔹 Visualización de un ejemplo aleatorio con sus valores de píxeles

```python
idx = np.random.randint(0, x_train.shape[0])
plt.imshow(x_train[idx])
plt.title(f"Clase: {class_names[y_train[idx][0]]}")
plt.colorbar()
plt.show()

print("Valores de píxeles (min-max):", x_train[idx].min(), "-", x_train[idx].max())
```
